```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d, CubicSpline, BarycentricInterpolator

# Given data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 2, 1, 3, 7, 4])

# New x values for interpolation
x_new = np.linspace(0, 5, 100)

# Linear interpolation
linear_interp = interp1d(x, y, kind='linear')
y_linear = linear_interp(x_new)

# Polynomial interpolation (using BarycentricInterpolator for simplicity)
poly_interp = BarycentricInterpolator(x, y)
y_poly = poly_interp(x_new)

# Cubic spline interpolation
cubic_spline_interp = CubicSpline(x, y)
y_cubic_spline = cubic_spline_interp(x_new)

# Nearest neighbor interpolation
nearest_interp = interp1d(x, y, kind='nearest')
y_nearest = nearest_interp(x_new)

# Plotting
plt.figure(figsize=(12, 8))

# Original data points
plt.scatter(x, y, color='red', label='Data Points')

# Linear interpolation plot
plt.plot(x_new, y_linear, label='Linear Interpolation', linestyle='--')

# # Polynomial interpolation plot
# plt.plot(x_new, y_poly, label='Polynomial Interpolation', linestyle='-.')

# # Cubic spline interpolation plot
# plt.plot(x_new, y_cubic_spline, label='Cubic Spline Interpolation', linestyle=':')

# # Nearest neighbor interpolation plot
# plt.plot(x_new, y_nearest, label='Nearest Neighbor Interpolation', linestyle='-')

# Plot customization
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Interpolation')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
measured_height = [2.35, 4.69, 9.63, 14.55, 20.10, 24.82]
percent_error = [2.22, 2.18, 1.85, 1.66, 0.83, 0.08]

# Plotting
plt.figure(figsize=(10, 6), dpi=200)
plt.plot(measured_height, percent_error, marker='o', label='data acquired')
plt.yscale('log')

# Adding titles and labels
plt.xlabel('Measured Height (mm)', fontsize=20)
plt.ylabel('% Error', fontsize=20)
plt.legend()

# Increase font size of axis ticks
plt.tick_params(axis='both', which='major', labelsize=20)

# Show plot
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()
```
